```python
#Measures of central tendency, dispersion and skewness for continuous variables.
NumericVars = Default_data._get_numeric_data().columns
DB = pd.DataFrame([],columns=['VariableName','Mean',"Median","Standard_deviation","Skewness","Kurtosis"],dtype="float")
DB.VariableName = DB.VariableName.astype('str')
DB.info()
DB
```

```python
NumericVars
```

```python
for position,var in enumerate(NumericVars):
    print(position,var)
```

```python
from scipy.stats import kurtosis
 
for position,var in enumerate(NumericVars):
    avg = round(Default_data[var].mean(),2)
    med = round(Default_data[var].median(),2)
    stdev = round(np.std(Default_data[var]),2)
    skewness = round(Default_data[var].skew(),2)
    Kurtosis = round(kurtosis(Default_data[var], fisher=True),2)
    x =pd.DataFrame({'VariableName':[var],'Mean':[avg],"Median":[med],
                     "Standard_deviation":[stdev],
                     "Skewness":[skewness],"Kurtosis":[Kurtosis]})
    DB = pd.concat([DB,x],axis=0)
```

```python
DB.reset_index()
# describe() does not give skewness and kurtosis.
# Used for loop and used skewness & kurtosis functions iteratively thro' all variables to study skewness & kurtosis.
# If kurtosis value is -ve then it has low peakedness than normal distribution.
# Credit amount has Kurtosis 4.27 is +ve means peakedness is higher than what normal distribution will have.
# If skewness value is +ve means it is positively skewed data. Cedit Amount skewness is 1.95 which is +ve ly skewed data
# If skewness value is -ve means it is negatively skewed data.
```

```python
## Evaluating Outliers with Z-score: Defining Python code to identify outliers with z-statistic for continuous variables.
# Analyzing Age variable distribution

import numpy as np
Age_mean = np.nanmean(Default_data.Age)
Age_stdev =np.std(Default_data.Age)
 
Default_data['Z_score_age'] = (Default_data.Age - Age_mean)/Age_stdev
 
Default_Data_Age_outliers = Default_data[(Default_data['Z_score_age']>3) | (Default_data['Z_score_age'] < -3)]
Default_Data_Age_outliers

# Any observation that goes beyond +3 and -3 standard deviation, flag it as an outlier.
```
