```python
# 4> Order the complaint types based on the average ‘Request_Closing_Time’, grouping them for different locations.
# Location ie City has null values. Filling all missing value with some default value here as "unknown".
data_nyc['City'].fillna('Unknown', inplace=True)
data_nyc['City'].head(50)
# Above in quesion 2> we have created column ‘Request_Closing_Time’.
# Now we want ‘Request_Closing_Time’ in Hr.
# Function to convert Time in Hour
def toHr(sh):
    days = sh.days
    hours = round(sh.seconds/3600, 2)
    result = (days * 24) + hours
    return result
data_nyc['Request_Closing_In_Hr'] = data_nyc['Request_Closing_Time'].apply(toHr)

data_nyc['Request_Closing_In_Hr'].head()
data_nyc.head(5)
```

```python
# 5> Perform a statistical test for the following:
# state the Null and Alternate and then provide a statistical test to accept or reject the Null Hypothesis along with the 
# corresponding ‘p-value’.
# 5.1> Whether the average response time across complaint types is similar or not (overall)
# I will use one-way ANOVA.
H_null  = 'all complaint types response time mean similar'
H_alt  = 'atleast one complaint types response time mean is significantly different.'

top7 = data_nyc['Complaint Type'].value_counts()[:7]
#h=['Complaint Type count']
top7 = pd.DataFrame(top7)
top7=top7.reset_index()
top7.rename(columns={'index':'Complaint Type','Complaint Type':'count'}, inplace = True)
top7
```

```python
grouped3 = data_nyc.groupby(['Complaint Type'])['Request_Closing_In_Hr'].agg('mean')
#grouped3[(grouped3.index==['Blocked Driveway','Illegal Parking'])]
grouped3 = pd.DataFrame(grouped3)
rowData = grouped3.loc[ ['Blocked Driveway' , 'Illegal Parking','Noise - Street/Sidewalk','Noise - Commercial','Derelict Vehicle','Noise - Vehicle','Animal Abuse'] , : ]
rowData=rowData.reset_index()
rowData
```

```python
bigdata=pd.merge(top7, rowData, how ='inner', on ='Complaint Type') 
bigdata.rename(columns={'Complaint Type':'Complaint_Type'}, inplace = True)
bigdata
```

```python
# Let's visualize data with boxplot
bigdata.boxplot('Request_Closing_In_Hr',by='Complaint_Type')
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
shdata=data_nyc
shdata=shdata.rename(columns={'Complaint Type':'Complaint_Type'})
#mod = ols('Request_Closing_In_Hr ~ Complaint_Type', data=bigdata).fit()
mod=ols('Request_Closing_In_Hr ~ Complaint_Type', data=shdata).fit()                
aov_table = sm.stats.anova_lm(mod)
print(aov_table)
mod.summary()
```

```python
# Our p-value of 0.0 is smaller than 0.05, so we reject our null hypothesis and we can accept our alternative. 
# ie. atleast one complaint types response time mean is significantly different.
# Even though the sample means of complaint type looks same, we do have a statistically significant difference.
# This means that with the observed data, there is enough evidence to assume a general difference in response times of 
# the 24 complaint types.
```
