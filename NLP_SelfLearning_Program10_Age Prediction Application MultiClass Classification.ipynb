```python
from nltk.tokenize.casual import EMOTICON_RE
```

# Multiclass classifier
step1:- which features r relevant to the classification task and using them to create feature set. Defined feature extraction function below which determines dictionary for each feature we include. These dictionaries r the feature sets and they r used to map feature names to their values.
Messages r collected from age specific chat rooms where each chat room is intended for users either they r in their 20s,30s or 40s. Will use this information to train a classifier that can help us predict the age of the user by looking at message they have sent.

```python
# defining feature extraction routine containing our 2 features.
# expected output is a dictionary showing features and their values.

def age_features(message):
    emoticon = EMOTICON_RE.findall(' '.join(message))
    length = len(message)
    if not emoticon:
        emoticon = 0 #emoji not present
    else:
        emoticon = 1 #emoji present
    return {'Emoticons':emoticon, 'Length': length}
```

```python
 # importing dataset
    
from nltk.corpus import nps_chat
chatroom20 = nps_chat.posts('10-19-20s_706posts.xml')
chatroom30 = nps_chat.posts('10-19-30s_705posts.xml')
chatroom40 = nps_chat.posts('10-19-40s_686posts.xml')
```

```python
chatroom40[54] #54 is message number. Each msg is list of the words it contains.
```

In order to train our classifier we need to label each of the messages with appropriate age indicator. Also will combine all of the data into single tuple labeled_chats.

```python
labeled_chats = ([(post, '20s') for post in chatroom20]+
                [(post, '30s') for post in chatroom30]+
                [(post, '40s') for post in chatroom40])
```

```python
labeled_chats[706] 
# output is tuple containing msg with age label
```

```python
# we r shuffling our labeled dataset so that we can split in train n test sets and then will map our features onto each msg
# in the labeled datasets.

import random
from sklearn.model_selection import train_test_split

random.shuffle(labeled_chats)
featuresets = [(age_features(n), age) for (n, age) in labeled_chats]
train_set, test_set = train_test_split(featuresets, test_size=0.2)
```

```python
train_set
# output is list of tuples.1 tuple for each msg in the chat dataset. Each tuple contains a dictionary of the features and 
# it's values for the specific msg as well as the age label we defined earlier. 
```

```python
# defining a classifier and train it onto dataset

import nltk
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
print(nltk.classify.accuracy(classifier, test_set)) # finally testing onto test set
```
