```python
# map lowercase string with our list of words tokens
single_tokenized_lowered=list(map(str.lower,word_tokenize(data2)))
print(single_tokenized_lowered)
```

```python
# Let's remove the stopwords using english stopwords list in nltk 
stopwords_en=set(stopwords.words('english'))
# by removing all stopwords from text we obtained
print([word for word in single_tokenized_lowered if word not in stopwords_en])
```

```python
# we also need to evaluate POS tag for each token
stop_words=set(stopwords.words('english'))
tokenized=sent_tokenize(data2)
# create a new wordlist and store the list of word tokens against each of the sentence tokens in data2
for i in tokenized:
    wordlist=nltk.word_tokenize(i)
# also we will check if there r any stopwords in recently created list
    wordlist=[w for w in wordlist if not w in stop_words]
# we will now tag the word tokens accordingly using the POS tags and print the tag output.
    tagged=nltk.pos_tag(wordlist)
    print(tagged)
```

```python
# for final text processing task we will be applying Named Entity Recognition to classify named entities in text into
# predefined categories such as the names of the persons, organizations, locations, expressions of times, quantities,
# monetary values, percentages etc
sentences=nltk.sent_tokenize(data2)
tokenized_sentences=[nltk.wordpunct_tokenize(sentence) for sentence in sentences]
tagged_sentences=[nltk.pos_tag(sentence) for sentence in tokenized_sentences]
# now pressed tagged_sentences under the chunk parser. If we set the parameter binary=true then named entities r just tagged
# as 'ne', otherwise classifier adds category label such as person,organization and gpe.
chunked_sentences=nltk.ne_chunk_sents(tagged_sentences,binary=True)
# create function along with empty list named as entity_names.
def extract_entity_names(t):
    entity_names=[]
    # we will now extract named entities from nltk chunked expression and store in empty created above
    if hasattr(t,'label') and t.label:
        if t.label()=='NE':
            entity_names.append(' '.join([child[0] for child in t]))
        else:
            for child in t:
                entity_names.extend(extract_entity_names(child))
    return entity_names

# again we will set entity_names as an empty list and we will extract entity names by iterating over each tree in chunked
# sentences
entity_names=[]
for tree in chunked_sentences:
    entity_names.extend(extract_entity_names(tree))

print(set(entity_names))
            
```

DESCRIPTION

The wiki corpus contains the full text of Wikipedia, and it contains 1.9 billion words in more than 4.4 million articles. You are provided with a subset of the same.
Objective:
• Implement different types of tokenizers
• Perform stemming and lemmatization
• Evaluate POS tags for the tokens
• Remove the stop words and print NER words

```python
# Step 1: Data Import and Tokenization

import nltk
with open('C:/Users/sharv/Documents/Post Graduate Program In Data Science-PurdueSimpliLearn/Course3-Machine Learning/Demo Datasets/Lesson 11/wiki_corpus.txt', 'r') as myfile:
    data=myfile.read().replace('\n', '')
for i, line in enumerate(data.split('\n')):
    if i> 10: # Lets take a look at the first 10 ads.
        break
    print(str(i) + ':\t' + line)
```

```python
from nltk import sent_tokenize, word_tokenize
sent_tokenize(data)
for sent in sent_tokenize(data):
    print(word_tokenize(sent))
```

```python
# Step 2: Stemming and Lemmatization

single_tokenized_lowered= list(map(str.lower, word_tokenize(data)))
from nltk.stem import PorterStemmer
porter = PorterStemmer()
for word in single_tokenized_lowered:
    print(porter.stem(word))
```

```python
from nltk.stem import WordNetLemmatizer
wnl= WordNetLemmatizer()
for word in single_tokenized_lowered:
    print(wnl.lemmatize(word))
```

```python
# Step 3: Evaluate POS Tags

stop_words= set(stopwords.words('english'))
tokenized = sent_tokenize(data)
for i in tokenized:
    wordsList= nltk.word_tokenize(i)
    wordsList= [w for w in wordsList if not w in stop_words]
    tagged = nltk.pos_tag(wordsList)
    print(tagged)
```

```python
# Step 4: Remove Stop Words and NER

def extract_entity_names(t):
    entity_names= []
    if hasattr(t, 'label') and t.label:
        if t.label() == 'NE':
            entity_names.append(' '.join([child[0] for child in t]))
    else:
        for child in t:
            entity_names.extend(extract_entity_names(child))
    return entity_names
```

```python
# Step 4: Remove Stop Words and NER (Contd.)

with open('C:/Users/sharv/Documents/Post Graduate Program In Data Science-PurdueSimpliLearn/Course3-Machine Learning/Demo Datasets/Lesson 11/wiki_corpus.txt', 'r') as f:
    for line in f:
        sentences = nltk.sent_tokenize(line)
        tokenized_sentences= [nltk.word_tokenize(sentence) for sentence in sentences]
        tagged_sentences= [nltk.pos_tag(sentence) for sentence in tokenized_sentences]
        chunked_sentences= nltk.ne_chunk_sents(tagged_sentences, binary=True)
        entities = []
        for tree in chunked_sentences:
            entities.extend(extract_entity_names(tree))
list_set= set(entities)
unique_list= (list(list_set))
for x in unique_list:
    print (x)
```
